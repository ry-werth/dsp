```python
import numpy as np

import nsfg
import first
import thinkstats2
import thinkplot
```

```python
r = np.random.random_sample((1000,))
```

```python
random_cdf = thinkstats2.Cdf(r, label='random_nums')
```

```python
thinkplot.Cdf(random_cdf)
thinkplot.Config(xlabel='Random Number', ylabel='CDF')
```

### ^^ As seen above the distribution is uniform. 20% of the numbers are less than .2, 40% are less than .4, 60% are less than .6 etc.

```python
random_pmf = thinkstats2.Pmf(r, label='random_nums')
```

```python
thinkplot.Pmf(random_pmf)
thinkplot.Config(xlabel='random number', ylabel='Pmf')
```

### ^^ As seen above the distribution is again uniform. All the values share a .1% (1/1000) probability.
